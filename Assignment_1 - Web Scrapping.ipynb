### Q1 - Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
wikipage = requests.get('https://en.wikipedia.org/wiki/Main_Page')
wikipage.content
soup = BeautifulSoup(wikipage.content,'html.parser')

# Create an empty list to store the header
header = []

# Code to get the header of the webpage
wikihead = soup.find_all('span', class_='mw-headline')

# Write a code to iterate over the tags extracted above
for i in wikihead:
    header.append(i.get_text())

import pandas as pd
df = pd.DataFrame({})
df['Header'] = header
df['S.No'] = list(range(1,len(header)+1))
df.set_index('S.No')
```

```python
df.to_csv('wiki.csv')
```

### Q3 - Write a python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.

```python
page = requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
page.content
soup = BeautifulSoup(page.content,'html.parser')

name = []
ratings = []
rank = []
release_year = []

# Name
names = soup.find_all('td',class_='titleColumn')
for i in range(0,len(names)):
    name1 = names[i]    #selecting one by one and appending names to the list
    name.append(name1.a.get_text())
    
# Release Year
year = soup.find_all('span',class_='secondaryInfo')
for i in year:
    release_year.append(i.get_text())

# Cleaning the data for this list and storing in right format
n_rel_year = []
for i in release_year:
    a = i.replace('(','')
    b = a.replace(')','')
    n_rel_year.append(int(b))
    
# IMDB Ratings
rate = soup.find_all('td',class_='ratingColumn imdbRating')
for i in range(0,len(rate)):
    rates = rate[i]
    ratings.append(float(rates.strong.get_text()))
    
# Rank 
rank = list(range(1,101))

# let's form a dataframe for the data
df = pd.DataFrame({})
df['Names'] = name[:100]
df['Rank'] = rank
df['Year of Release'] = n_rel_year[:100]
df['IMDB Ratings'] = ratings[:100]
```

```python
# Finalising the model
df.set_index('Rank',inplace=True)
df.head(10)
```

### Q4 - Write a python program to scrap book name, author name, genre and book review of any 5 books from ‘www.bookpage.com’

```python
# Requesting data from the URL
bookpage = requests.get('https://bookpage.com/reviews?page=1')
bookpage.content

# Parsing the data
soup = BeautifulSoup(bookpage.content,'html.parser')

# Creating the empty lists of required data
author = []
novel = []
genre = []
review = []

# Author
auth = soup.find_all('p',class_='sans bold')
for i in auth:
    author.append(i.get_text().replace('\n',''))
    
# Novel
nov = soup.find_all('div',class_='flex-article-content')
for i in range(0,len(nov)):
    novels = nov[i]
    novel.append(novels.h4.a.get_text())
    
# Cleaning this list
new_novel = []
for i in novel:
    a = i.replace('★ ','')
    new_novel.append(a)
    
# Genre
gen = soup.find_all('p',class_='genre-links hidden-phone')
for i in gen:
    a = i.get_text().split()
    b = ' '.join([str(ele) for ele in a]) 
    genre.append(b)
    

revw = soup.find_all('p',class_='excerpt')
for i in revw:
    review.append(i.get_text().replace('\n',''))

# Creating a dataframe
df = pd.DataFrame({})
df['Author'] = author[:5]
df['Novel'] = new_novel[:5]
df['Genre'] = genre[:5]
df['Review'] = review[:5]
df
```

```python
df.to_csv('bookpage.csv')
```

### Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape: 

i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating. 

```python
#We will request the URL to get the data 
page = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page.content

#We will parse the data using beautifulsoup
soup = BeautifulSoup(page.content,'html.parser')

#creating lists of required data

team = []
matches = []
points = []
ratings = []

# Name
nam = soup.find_all('span',class_='u-hide-phablet')
for i in nam:
    team.append(i.get_text())
    
# Matches(For England)
mat = soup.find('td',class_='rankings-block__banner--matches')
matches.append(int(mat.get_text()))

# Points(For England)
point = soup.find('td',class_='rankings-block__banner--points')
points.append(int(point.get_text().replace(',','')))

mats = soup.find_all('td',class_='table-body__cell u-center-text')
ls_matches = mats[::2]
ls_points = mats[1::2] 
for i in ls_matches: 
    matches.append(int(i.get_text()))
for m in ls_points:
    points.append(int(m.get_text().replace(',','')))
    
# Ratings(For England)
rat = soup.find('td',class_='rankings-block__banner--rating u-text-right')
ratings.append(int(rat.get_text().strip()))

# Ratings (for rest of the team)
rats = soup.find_all('td',class_='table-body__cell u-text-right rating')
for i in rats:
    ratings.append(int(i.get_text()))
    
# Creating a dataframe
df = pd.DataFrame({})
df['Team'] = team[:10]
df['Match'] = matches[:10]
df['Point'] = points[:10]
df['Ratings'] = ratings[:10]

df
```

### Q9 - Write a python program to scrape ‘software developer’ job listings from ‘Monster.com’. It should include all the jobs listed for the next 5 pages.

```python
def monster(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    jobs=[]
    job=soup.find_all('h2',class_='title')
    for i in job:
        jobs.append(i.text.replace('\r\n',''))
    
    Days=[]
    day=soup.find_all('div',class_='meta flex-col')
    for i in day:
        Days.append(i.text.replace('\n','').replace('Applied\n\n\n\nSaved\n\n',''))
    
    Company=[]
    comp=soup.find_all('span',class_='name')
    for i in range(0,len(comp),2):
        Company.append(comp[i].text)
        
    Location=[]
    for i in range(1,len(comp),2):
        Location.append(comp[i].text.replace('\r\n',''))
        
    import pandas as pd
    monster=pd.DataFrame({})
    monster['jobs']=jobs
    monster['Days']=Days
    monster['Company']=Company
    monster['Location']=Location
    return(monster)
```

```python
df=monster('https://www.monster.com/jobs/search/?q=Software-Developer&where=india')
df.head(10)
```
